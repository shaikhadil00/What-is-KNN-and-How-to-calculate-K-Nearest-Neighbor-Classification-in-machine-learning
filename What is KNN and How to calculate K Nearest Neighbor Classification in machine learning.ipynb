```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize the KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)  # You can choose the number of neighbors
```

```python
# Train the model
knn.fit(X_train, y_train)
```

```python
# Make predictions on the testing data
y_pred = knn.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# Assuming X_train and X_test are already loaded
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize the KNN classifier with "ball_tree" algorithm
knn_ball_tree = KNeighborsClassifier(n_neighbors=3, algorithm='ball_tree')

# Train the model
knn_ball_tree.fit(X_train, y_train)
```

```python
# Make predictions on the testing data
y_pred_ball_tree = knn_ball_tree.predict(X_test)

# Evaluate the model (optional)
accuracy_ball_tree = accuracy_score(y_test, y_pred_ball_tree)
print(f"Accuracy (ball_tree): {accuracy_ball_tree:.2f}")
```

```python
# Create a scatter plot using petal length and sepal length
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 2], hue=y_test, palette='viridis')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.title("Scatter Plot: Sepal Length vs. Petal Length (ball_tree)")
plt.legend(title="Species")
plt.show()
```
